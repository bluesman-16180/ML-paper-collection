```python
"""
Author: Vishal Keshav (bulletcross)
"""

import matplotlib.pyplot as plt
import numpy as np
from numpy import ma
```

```python
v1 = [1,2]
v2 = [-2,4]
v3_init = [-4,-2]
v3_end = [1,2]

vectors = [v1,v2]

fig = plt.figure(figsize = (10, 10))                                                               
ax = fig.add_subplot(1,1,1)

major_ticks = np.arange(-10,10,1)
ax.set_xticks(major_ticks)
ax.set_yticks(major_ticks)

plt.quiver(0,0,v1[0],v1[1], angles='xy',scale_units='xy', scale=1,color='r')
#plt.quiver(0,0,v2[0],v2[1], angles='xy',scale_units='xy', scale=1)
plt.quiver(v3_init[0],v3_init[1],v3_end[0],v3_end[1], angles='xy',scale_units='xy', scale=1,color='b')
#plt.quiver([0, 0, 0], [0, 0, 0], [1, -2, 4], [1, 2, -7], angles='xy', scale_units='xy', scale=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(which='both')
plt.show()
```

```python
#Transformation of a vector

my_vector = np.array([2,4])

#1. Scaling a vector

M = np.array([[2,0],[0,2]])
transformed_vector = M.dot(my_vector)
```

```python
#before scale transformation
fig = plt.figure()                                                               
ax = fig.add_subplot(1,1,1)
major_ticks = np.arange(-10,10,1)
ax.set_xticks(major_ticks)
ax.set_yticks(major_ticks)

plt.quiver(0,0,my_vector[0],my_vector[1], angles='xy',scale_units='xy', scale=1,color='r')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(which='both')
plt.show()
```

```python
#After scaling
plt.quiver(0,0,my_vector[0],my_vector[1], angles='xy',scale_units='xy', scale=1,color='r')
plt.quiver(0,0,transformed_vector[0],transformed_vector[1], angles='xy',scale_units='xy', scale=1,color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(which='both')
plt.show()
```

```python
#After translation
higher_dim_my_vector = [2,4,1]
higher_dim_M = np.array([[1,0,2],[0,1,0],[0,0,1]])
higher_dim_transformed_vector = higher_dim_M.dot(higher_dim_my_vector)
transformed_vector = higher_dim_transformed_vector[0:2]

plt.quiver(0,0,my_vector[0],my_vector[1], angles='xy',scale_units='xy', scale=1,color='r')
plt.quiver(0,0,transformed_vector[0],transformed_vector[1], angles='xy',scale_units='xy', scale=1,color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(which='both')
plt.show()
```

```python
#After rotation
angle = 100.*np.pi / 180.0
M = np.array([[np.cos(angle), -np.sin(angle)],[np.sin(angle),np.cos(angle)]])
transformed_vector = M.dot(my_vector)
plt.quiver(0,0,my_vector[0],my_vector[1], angles='xy',scale_units='xy', scale=1,color='r')
plt.quiver(0,0,transformed_vector[0],transformed_vector[1], angles='xy',scale_units='xy', scale=1,color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(which='both')
plt.show()
```

```python
#After reflection along x axis
M = np.array([[1, 0],[0,-1]])
transformed_vector = M.dot(my_vector)
plt.quiver(0,0,my_vector[0],my_vector[1], angles='xy',scale_units='xy', scale=1,color='r')
plt.quiver(0,0,transformed_vector[0],transformed_vector[1], angles='xy',scale_units='xy', scale=1,color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(which='both')
plt.show()
```

```python
#Lets take a random linear transformation matrix
""" 
| 5  -2 |
| 3   4 |

"""

M = np.array([[5,2],[3,4]])
vector1 = np.array([1,-2])
special_vector1 = np.array([1,1])
special_vector2 = np.array([2,-3])

transformed_vector = M.dot(vector1)
plt.figure(1)
#plt.subplot(331)
plt.quiver(0,0,vector1[0],vector1[1], angles='xy',scale_units='xy', scale=1,color='r')
plt.quiver(0,0,transformed_vector[0],transformed_vector[1], angles='xy',scale_units='xy', scale=1,color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(which='both')
plt.show()

transformed_vector = M.dot(special_vector1)
plt.figure(2)
#plt.subplot(331)
plt.quiver(0,0,special_vector1[0],special_vector1[1], angles='xy',scale_units='xy', scale=1,color='r')
plt.quiver(0,0,transformed_vector[0],transformed_vector[1], angles='xy',scale_units='xy', scale=1,color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(which='both')
plt.show()

transformed_vector = M.dot(special_vector2)
plt.figure(3)
#plt.subplot(331)
plt.quiver(0,0,special_vector2[0],special_vector2[1], angles='xy',scale_units='xy', scale=1,color='r')
plt.quiver(0,0,transformed_vector[0],transformed_vector[1], angles='xy',scale_units='xy', scale=1,color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(which='both')
plt.show()
```
